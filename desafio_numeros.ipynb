```python
#verificar par ou impar 

lista_numeros = []

while True:

    numero = int(input('Informe um numero ou "Q" para SAIR'))
    resultado = int(numero % 2)
      
    if numero % 2 == 0:
            lista_numeros.append(numero)
            print (f'{numero} é par')

    elif numero % 2 != 0:
            lista_numeros.append(numero)
            print (f'{numero} é impar')

    else:
           break    
    print(lista_numeros)
```
